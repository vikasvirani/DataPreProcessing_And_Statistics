```python
%matplotlib inline

#Task 1: Data Preparation
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pandas.plotting import scatter_matrix

# Load the data
Car_data = pd.read_csv("Automobile.csv",sep="#",header=None,names=["symboling","normalized-losses","make","fuel-type",
                        "aspiration","num-of-doors","body-style","drive-wheels","engine-location","wheel-base","length",
                        "width","height","curb-weight","engine-type","num-of-cylinders","engine-size","fuel-system","bore",
                        "stroke","compression-ratio","horsepower","peak-rpm","city-mpg","highway-mpg","price"])

print('Displaying Sample records')
Car_data.head()

print('Displaying shape of DataFrame')
Car_data.shape

#Checking Data types
print('Displaying data types of all columns')
Car_data.dtypes

#Car_data['symboling'] = Car_data['symboling'].astype('object')

#Car_data.dtypes


#Checking for Typos
columns = ['fuel-type','num-of-doors','body-style','drive-wheels','engine-location','num-of-cylinders','fuel-system','aspiration',
           'make','engine-type']

print('Displaying Value Counts')
for x in columns:
    print('Value Counts for : '+x)
    Car_data[x].value_counts()


#Removing Typos
for x in columns:
    Car_data[x] = Car_data[x].str.lower()
    

def replace_string(x,actual,replaced):
    ret = ''
    if (x == actual):
        ret = replaced
    else:
        ret = x
        
    return ret


Car_data['num-of-doors'] = Car_data['num-of-doors'].apply(lambda x: replace_string(x,'fourr','four'))
Car_data['aspiration'] = Car_data['aspiration'].apply(lambda x: replace_string(x,'turrrrbo','turbo'))
Car_data['make'] = Car_data['make'].apply(lambda x: replace_string(x,'vol00112ov','volvo'))


#Removing for Redundant Whitspace
for x in columns:
    Car_data[x] = Car_data[x].str.strip()
    
    
#Checking final Value Counts
print('Displaying Final Value Counts')
for x in columns:
    print('Value Counts for : '+x)
    Car_data[x].value_counts()

        
#Sanity Checks
check_price = Car_data['price'] != 0.0
Car_data = Car_data[check_price]

Car_data['symboling'].value_counts()
Car_data['symboling'] = Car_data['symboling'].apply(lambda x: replace_string(x,4,0))


#Missing Values & Outliers
print('Displaying Missing Values')
Car_data.isnull().sum()

Car_data['normalized-losses'].fillna(Car_data['normalized-losses'].mean(axis=0), inplace=True)
Car_data['bore'].fillna(Car_data['bore'].mean(axis=0), inplace=True)
Car_data['stroke'].fillna(Car_data['stroke'].mean(axis=0), inplace=True)
Car_data['horsepower'].fillna(Car_data['horsepower'].mean(axis=0), inplace=True)
Car_data['peak-rpm'].fillna(Car_data['peak-rpm'].mean(axis=0), inplace=True)
Car_data['price'].fillna(Car_data['price'].mean(axis=0), inplace=True)

Car_data['num-of-doors'].fillna('four', inplace=True)

print('Displaying Box plot for Outliers');
sns.boxplot(y=Car_data['normalized-losses'], palette="cubehelix")
plt.title('Normalized Losses')
plt.show()
```

```python
# Task 2: Data Exploration
# 1. Choose 1 column with nominal values, 1 column with ordinal Values, and 1 column with numerical values

#Nominal Column : 'make'
Car_data['make'].value_counts().plot(kind='bar')
plt.title('Number of Cars by Each Make');
plt.xlabel('Car Make');
plt.ylabel('Count');
plt.show();

#Ordinal Column : 'num-of-cylinders'
Car_data['num-of-cylinders'].value_counts().plot(kind='pie',autopct='%.2f',figsize=(10, 10))
plt.title('Number of Cylinders in Car');
plt.ylabel('');
plt.show();

#Numeric Column : 'Price'
Car_data['price'].plot(kind='density')
plt.title('Price Distribution for Cars');
plt.xlabel('Price');
plt.ylabel('Density');
plt.show();
```

```python
# Task 2: Data Exploration
# 2. Explore the relationships between columns; at least 3 visualisations with plausible hypothesis

#Hypothesis - 1 : Do city fuels economy increases as highway fuel economy increases in the car?
Car_data.plot(kind='scatter',x='city-mpg',y='highway-mpg',color=['orange'])
plt.title('City Vs. Highway Fuel Economy');
plt.xlabel('City Economy');
plt.ylabel('Highway Economy');
plt.show();

#Hypothesis - 2 : Do the average price of car tends to be higher based on make?
df = Car_data.groupby('make')['price'].mean().round(2)
df = pd.DataFrame(df).sort_values('price')
df.plot(kind='bar',color=['cyan'])
plt.title('Average price by Car Make');
plt.xlabel('Car Make');
plt.ylabel('Average Price');
plt.show();

#Hypothesis - 3 : Can Aspiration influence the horsepower of the Car?
# Use a color palette
sns.boxplot(x=Car_data["aspiration"], y=Car_data["horsepower"], palette="cubehelix")
plt.title('Horsepower by Aspiration');
plt.xlabel('Aspiration');
plt.ylabel('Horsepower');
plt.show();
```
